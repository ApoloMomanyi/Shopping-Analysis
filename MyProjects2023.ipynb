```python
"""
A graph showing the Spending Score based on Family Size
"""
x=pd.read_csv('C:/My python files/Customers.csv')
x.set_index("CustomerID",inplace=True)
x.groupby("Family Size")["Spending Score (1-100)"].mean().plot()
plt.xlabel("Family size")
plt.ylabel("Spending Score")
plt.suptitle("Spending Score Based on Family Size")
```

```python
df=pd.read_csv('C:/My python files/Customers.csv')
df.set_index("CustomerID",inplace=True)
df.groupby("Family Size")["Annual Income ($)"].mean().plot(kind="area",stacked=False)
plt.xlabel("Family Size")
plt.ylabel("Annual Income")
plt.suptitle("Annual Income and Family Size")
plt.savefig("output.png")
plt.show()
```
